# Plotly-Only Table2Charts Project

This notebook is a lightweight analysis entrypoint for the course project scaffold. It is intended for reading normalized experiment outputs, making tables and plots, and documenting conclusions. It is not intended to launch full GPU training by default.


## Project Context

- Original Excel corpus is unavailable, so experiments use the public Plotly-based adaptation.
- Current experiment families include SFT/original greedy, RL greedy, updated policy, dense reward, dense reward plus updated policy, and actor-critic diagnostics.
- Heavy training and evaluation should be launched on the remote GPU host, not from this notebook.


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

repo_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
metrics_path = repo_root / 'experiments' / 'results' / 'metrics.csv'
df = pd.read_csv(metrics_path)
df.head()
```

```python
if not df.empty:
    display(df[['method', 'sampling_strategy', 'reward_mode', 'score_mode', 'R@1', 'R@3', 'R@5', 'R@10', 'R@20', 'recall_all']].sort_values(['method', 'sampling_strategy']))
else:
    print('metrics.csv is currently empty except for the header; run the collector first.')
```

```python
epsilon_df = df[df['sampling_strategy'] == 'epsilon_top_m'].copy()
if not epsilon_df.empty:
    epsilon_df['epsilon_start'] = pd.to_numeric(epsilon_df['epsilon_start'], errors='coerce')
    epsilon_df['R@1'] = pd.to_numeric(epsilon_df['R@1'], errors='coerce')
    epsilon_df = epsilon_df.sort_values('epsilon_start')
    ax = epsilon_df.plot(x='epsilon_start', y='R@1', marker='o', title='Epsilon Sweep: R@1')
    ax.set_xlabel('epsilon_start')
    ax.set_ylabel('R@1')
    plt.show()
else:
    print('No epsilon sweep rows are available yet.')
```
